```python
minutes = []
hours = []
seconds = []
# Looping for digunakan untuk mengakses setiap elemen waktu tersebut.
for t in df.Time:# df.Time merupakan kolom waktu dalam sebuah DataFrame pada Pandas.
    t  = t.split(":")
    minutes.append(t[1])
    hours.append(t[0])
    seconds.append(t[2])
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['Jam'] = df['Time'].dt.hour
df['Menit'] = df['Time'].dt.minute
df['Detik'] = df['Time'].dt.second
```

```python
df.drop(['Time'], axis=1, inplace=True)
df.head()
```

MB (Body Wave Magnitude): Skala ini mengukur magnitudo berdasarkan gelombang P (primer) dan S (sekunder) yang merambat melalui inti bumi. MB biasanya digunakan untuk mengukur magnitudo gempa bumi kecil hingga menengah.

MS (Surface Wave Magnitude): Skala ini mengukur magnitudo berdasarkan amplitudo gelombang permukaan yang merambat di sepanjang kerak bumi. MS digunakan untuk mengukur magnitudo gempa bumi dengan jangkauan frekuensi menengah hingga rendah.

MWb (Moment Magnitude Body Wave): Skala ini mengukur magnitudo berdasarkan momen seismik yang dihasilkan oleh pergerakan lempeng. MWb merupakan perkiraan awal magnitudo menggunakan data gelombang body wave.

MWW (Moment Magnitude by Long-period Wave): Skala ini mengukur magnitudo berdasarkan momen seismik menggunakan data gelombang periode panjang. MWW memberikan perkiraan yang lebih akurat untuk magnitudo gempa bumi yang kuat.

MW (Moment Magnitude): Skala ini mengukur magnitudo berdasarkan momen seismik yang dihasilkan oleh pergerakan lempeng. MW adalah skala yang lebih umum digunakan dan memberikan perkiraan yang lebih akurat untuk magnitudo gempa bumi kuat.

MWC (Moment Magnitude by Regional Wave): Skala ini mengukur magnitudo berdasarkan momen seismik menggunakan data gelombang regional. MWC digunakan untuk mengukur magnitudo gempa bumi di wilayah tertentu.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time

t0=time.time()
clf_dt = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5,
                                min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0,
                                max_features=None, random_state=None, max_leaf_nodes=None,
                                min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
clf_dt.fit(X_train,y_train)
y_pred = clf_dt.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('accuracy :',score)
print('Time taken :' , time.time()-t0)
```

```python
import pandas as pd
from datetime import datetime, timedelta
from sklearn.tree import DecisionTreeClassifier
import random

# Load historical earthquake data
data = df
data

```

```python
# Generate predictions for the next 365 days
start_date = datetime(2022, 10, 5)
end_date = start_date + timedelta(days=365)
```

```python

date_range = pd.date_range(start=start_date, end=end_date, freq='D')

predictions = pd.DataFrame(columns=['Date', 'Prediction'])
```

```python
# Melatih model decision tree
model = DecisionTreeClassifier()
model.fit(X, y)

# Prediksi untuk setiap tanggal dalam rentang yang ditentukan
predictions = pd.DataFrame(columns=['Date', 'Prediction'])
date_range = pd.date_range(start_date, end_date)  # Ganti start_date dan end_date dengan rentang tanggal yang diinginkan

for date in date_range:
    # Memilih sampel acak dari data historis
    sample = data.sample(n=1)

    # Mengambil nilai dari setiap fitur dari sampel acak
    day = int(sample['Day'])
    month = int(sample['Month'])
    years = int(sample['Years'])
    latitude = sample['Latitude'].values[0]
    longitude = sample['Longitude'].values[0]
    depth = sample['Depth'].values[0]
    jam = int(sample['Jam'])
    menit = int(sample['Menit'])
    detik = int(sample['Detik'])

    # Melakukan prediksi menggunakan model yang dilatih
    prediction = model.predict([[day, month, years, latitude, longitude, depth, jam, menit, detik]])

    # Menambahkan hasil prediksi ke dalam DataFrame predictions
    predictions = predictions.append({
        'Date': date,
        'Prediction': prediction[0]
    }, ignore_index=True)

# Menampilkan hasil prediksi
print(predictions)
```

```python
predictions.to_csv('hasil_prediksi.csv', index=False)
```
